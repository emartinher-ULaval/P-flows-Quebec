# Milk

## Milk production for whole Quebec (Milk production statistics, by administrative region, regional county municipality (RCM) and for all of Québec, https://statistique.quebec.ca/en/document/milk-production-statistics-by-administrative-region-regional-county-municipality-rcm-and-for-all-of-quebec) and assigned to each Census Division based on the number of dairy cows reported by the Census of agriculture

```python
Milk_Quebec = 33478860*100 #liters

#Import animals in CDs from manure files. Dairy heifers do not produce milk
CattleManureQuebec = pd.read_csv('DataConditioning/'+'Cattle'+'_'+'Manure'+'.csv', sep=",", header=0)

DairyCowQuebec = CattleManureQuebec[['CD','DairyCow', 'DairyCow_AU']]

DairyCowTotal = DairyCowQuebec.sum(numeric_only=True)['DairyCow']
DairyCowAUTotal = DairyCowQuebec.sum(numeric_only=True)['DairyCow_AU']

#Share of cows in each CD
DairyCowQuebec['% of dairy cows (animals) in CD over total Quebec'] = DairyCowQuebec['DairyCow']/DairyCowTotal*100
DairyCowQuebec['% of dairy cows (AU) in CD over total Quebec'] = DairyCowQuebec['DairyCow_AU']/DairyCowAUTotal*100

# Assingning milk based on cows
DairyCowQuebec['Milk (liters)'] = DairyCowQuebec['% of dairy cows (animals) in CD over total Quebec']/100*Milk_Quebec

DairyCowQuebec
```

```python
P_Milk = 0.94E-3 #kg per L
```

```python
DairyCowQuebec['Milk (kg P)'] = DairyCowQuebec['Milk (liters)']*P_Milk

DairyCowQuebec.to_csv('../Results/AgriProducts/P_MilkQuebec.csv')

DairyCowQuebec
```

For rabbits we took the percentaje over the total rabbits slaughtered in Canada for 2019 (last year available) reported by provincial governemtn (19%, to account for a new facility oppened in 2019) (Portrait-diagnostic sectoriel de l'industrie cunicole au Québec
 https://www.mapaq.gouv.qc.ca/fr/md/Publications/Pages/Details-Publication.aspx?guid=%7b389dc98d-89de-4873-956b-37486ade2a04%7d) and applied to the total number of rabbit reported by the federal government for 2021 (https://agriculture.canada.ca/en/sector/animal-industry/red-meat-and-livestock-market-information/rabbit-industry-glance)

For poultry other than chicken, Quebec and the atlantic provinces are rported together. The data was broken down using the % of chicken slaughtered in Quebec regarding the total chicken slaughted in the sum of Quebec and the atlantic provinces (71.9%)

For cattle, the number of animals slaughtered in federal slaughterhouses is provided as the sum of Quebec and the atlantic provinces. For breaking the data down, the % of animals sluaghtered in ferderal slaughterhoused over the total of animals slaughtered in Quebec for 2019 (last year available) is taken from a provincial report (for calves 92.3% are federal and 7.7% provincial, and for cows 46.1% are federal and 53.9% are provincial), and then applied to the animals slaugthered in the provincial plants of Quebec

### Breaking down data into census division level

Considerations:

15% of total livestock is processed in provincially inspected facilities. It is assumed that 100% of sheep, lambs, goats, and rabbits are processed in these faciltities (https://www.attorneygeneral.jus.gov.on.ca/english/about/pubs/meatinspectionreport/Chapter13.pdf)

The remaining 85% of total livestock, comprised of cattle, swine, and poultry, is processed in federally inspected facilities.(https://www.attorneygeneral.jus.gov.on.ca/english/about/pubs/meatinspectionreport/Chapter13.pdf)

NOT Federally inspected facilities are geolocated based on the data provided by the Canadian Food Inspection Agency (https://inspection.canada.ca/active/scripts/meavia/reglist/reglist.asp) by using Nominatim and OpenStreetMaps. Only facilities located in Ontario reporting slaughteirng activity are considered.

NOT Geolocation of provincially inspected facilities is not provided by OMAFRA (https://data.ontario.ca/en/dataset/provincially-licensed-meat-plants). 

Therefore, the animals slaughtered in each census division are estimated based on the emplyoment data for meat processing plants reported by the Canadian Business Counts report for each year under evaluation since 2018 (because of this is the first year reported at census division level (https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3310039701&pickMembers%5B0%5D=3.168). The NAICS code for slaughtering is the 311611, but 3116 (meat processing) havs been used

```python
EmployeesMeatPlants = pd.read_csv('../Datasets/AgriProductsProduction/Meat/EmployeesMeatPlants/2021/33100397-eng/33100397.csv', sep=",", header=0)

# Filtering for Meat product manufacturing [3116] activities
EmployeesMeatPlants = EmployeesMeatPlants[EmployeesMeatPlants['North American Industry Classification System (NAICS)'].str.contains("3116")==True]

# Filtering for Quebec facilities
DGUID_Vintage = '2016' #Someone in the goverment forgot to upodate the vintage of the file
DGUID_Type = 'A'
DGUID_Schema =  '0005' #The data is reporte at census subdiviosn
DGUID_Quebec = '24'
DGUID_filter = DGUID_Vintage+DGUID_Type+DGUID_Schema+DGUID_Quebec

EmployeesMeatPlants_Quebec = EmployeesMeatPlants[EmployeesMeatPlants['DGUID'].str.contains(DGUID_filter)==True]

# Removing rows Total, with employees
EmployeesMeatPlants_Quebec = EmployeesMeatPlants_Quebec[~EmployeesMeatPlants_Quebec['Employment size'].str.contains("Total, with employees")]

# Obtaining  CD
EmployeesMeatPlants_Quebec = EmployeesMeatPlants_Quebec.reset_index(drop=True)
EmployeesMeatPlants_Quebec['CD'] = [i.split(DGUID_Vintage+DGUID_Type+DGUID_Schema)[1][0:4] for i in EmployeesMeatPlants_Quebec['DGUID'].to_list()]

#Estimating employees
EmployeesDict = {'1 to 4 employees':2.5,
                     '5 to 9 employees':7,
                        '10 to 19 employees':14.5,
                        '20 to 49 employees':34.5,
                        '50 to 99 employees':74.5,
                        '100 to 199 employees':149.5,
                        '200 to 499 employees':349.5,
                        '500 plus employees':700,
                    }

EmployeesMeatPlants_Quebec['Employees'] = np.nan
for index, row in EmployeesMeatPlants_Quebec.iterrows():
    EmployeesMeatPlants_Quebec.loc[index, 'Employees'] = EmployeesDict[EmployeesMeatPlants_Quebec.loc[index, 'Employment size']]*EmployeesMeatPlants_Quebec.loc[index,'VALUE'].astype(int)

# Grouping by Census Division
EmployeesMeatPlants_Quebec = EmployeesMeatPlants_Quebec.groupby(['CD']).sum(numeric_only=True).drop(['UOM_ID','SCALAR_ID','VALUE','STATUS','SYMBOL','TERMINATED','DECIMALS'], axis=1)

# Normalization
# EmployeesMeatPlants_Ontario['NormalizationIndex'] = (EmployeesMeatPlants_Ontario['Employees']-EmployeesMeatPlants_Ontario['Employees'].min())/(EmployeesMeatPlants_Ontario['Employees'].max()-EmployeesMeatPlants_Ontario['Employees'].min())
EmployeesMeatPlants_Quebec['NormalizationIndex(%)'] = (EmployeesMeatPlants_Quebec['Employees']/EmployeesMeatPlants_Quebec['Employees'].sum())*100

# Including those census diviosns with no facilities
CensusDivisions = pd.read_csv('../Datasets/GeodataFiles/CensusDivisions/georef-canada-census-division.csv', sep=";", header=0)
CensusDivisionsQuebec = CensusDivisions[CensusDivisions['Official Code Province / Territory'] == 24].reset_index(drop=True)
CensusDivisionsQuebec_CD=CensusDivisionsQuebec[['Official Code Census Division']].astype(str).rename(columns={'Official Code Census Division':'CD'})

EmployeesMeatPlants_Quebec = EmployeesMeatPlants_Quebec.reset_index()
EmployeesMeatPlants_Quebec = EmployeesMeatPlants_Quebec.merge(CensusDivisionsQuebec_CD, on='CD', how='outer', sort=True).fillna(0)

EmployeesMeatPlants_Quebec = EmployeesMeatPlants_Quebec.set_index('CD')

EmployeesMeatPlants_Quebec.to_csv('../Datasets/AgriProductsProduction/Meat/EmployeesMeatPlants/2021/FederallyLicensedMeatPlants_Quebec.csv')

EmployeesMeatPlants_Quebec
```

```python
# We convert everyting to multirow
iterables = [AnimalsTotal.columns.to_list(), AnimalsTotal.index.to_list()]
multiindex_AnimalsTotal_CD = pd.MultiIndex.from_product(iterables, names=["Year", "Animal"])


AnimalsTotal_CD = pd.DataFrame(index=['Total Quebec'], columns=multiindex_AnimalsTotal_CD)
for i in AnimalsTotal.columns.to_list():
    for ii in AnimalsTotal.index.to_list():
        AnimalsTotal_CD.loc['Total Quebec',(i,ii)] = AnimalsTotal.loc[ii,i]
        
# Adding Census divisons
for i in EmployeesMeatPlants_Quebec.index.to_list():
    AnimalsTotal_CD.loc[i,:] = AnimalsTotal_CD.loc['Total Quebec',:]*(EmployeesMeatPlants_Quebec.loc[i,'NormalizationIndex(%)']/100)

AnimalsTotal_CD.index.name = 'CD'       

AnimalsTotal_CD.to_csv('../Datasets/AgriProductsProduction/Meat/Conditioned/AnimalsTotal_CD.csv')

AnimalsTotal_CD
```

```python
AnimalsTotalCD_Final = AnimalsTotal_CD.copy()
AnimalsTotalCD_Final = AnimalsTotalCD_Final.drop('Total Quebec', axis=0)

AnimalsTotalCD_index = AnimalsTotal_CD.drop('Total Quebec', axis=0).index.to_list()
# AnimalsTotalCD_columns = [i[0].split(' (heads)')[0] for i in AnimalsTotal_CD.columns.to_list()]
AnimalsTotalCD_columns = AnimalsTotal_CD.drop('Total Quebec', axis=0).columns.to_list()

for i in AnimalsTotalCD_index:
    for ii in AnimalsTotalCD_columns:
        AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]  +' (kg live)',ii[1])] = (AnimalsTotalCD_Final.loc[i,ii]*CarcassWeight.loc[ii[1],ii[0].split(' (heads)')[0]]/
                                                              (LivetoCarcassWeight_ratio.loc[ii[1],'Carcass_Live_ratio (%)']/100))
        
        AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]  +' (kg carcass)',ii[1])] = (AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]   +' (heads)',ii[1])]*
                                                                                          CarcassWeight.loc[ii[1],ii[0].split(' (heads)')[0]])
        
        AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]  +' (kg slaugther waste)',ii[1])] = (AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]   +' (kg live)',ii[1])]-
                                                                 AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]   +' (kg carcass)',ii[1])])
        
        AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]  +' (kg P live)',ii[1])] = (AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0]  +' (kg live)',ii[1])]*
                                                                          AnimalComposition.loc[ii[1],'P (%live_wt)']/100)
        
        AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0] +' (kg P carcass)',ii[1])] = (AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0] +' (kg carcass)',ii[1])]*
                                                                          AnimalComposition.loc[ii[1],'P (%live_wt)']/100)
        
        AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0] +' (kg P slaugther waste)',ii[1])] = (AnimalsTotalCD_Final.loc[i,(ii[0].split(' (heads)')[0] +' (kg slaugther waste)',ii[1])]*
                                                                          AnimalComposition.loc[ii[1],'P (%live_wt)']/100)
                
        # AnimalsTotal_Final.loc[i,ii+' (kg slaugther waste)'] = (AnimalsTotal_Final.loc[i,ii +' (kg live)'] -
        #                                                         AnimalsTotal_Final.loc[i,ii +' (kg bones)'] - 
        #                                                         AnimalsTotal_Final.loc[i,ii +' (kg meat)'] -
        #                                                         AnimalsTotal_Final.loc[i,ii +' (kg fat)'] -
        #                                                         AnimalsTotal_Final.loc[i,ii +' (kg blood)'])
AnimalsTotalCD_Final.to_csv('../Datasets/AgriProductsProduction/Meat/Conditioned/AnimalsTotalCD_Final.csv')
AnimalsTotalCD_Final.to_csv('../Results/AgriProducts/P_MeatQuebecRaw.csv')


AnimalsTotalCD_Final
```
