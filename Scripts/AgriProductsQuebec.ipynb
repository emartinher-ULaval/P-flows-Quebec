# Milk

## Milk production for whole Quebec (Milk production statistics, by administrative region, regional county municipality (RCM) and for all of Québec, https://statistique.quebec.ca/en/document/milk-production-statistics-by-administrative-region-regional-county-municipality-rcm-and-for-all-of-quebec) and assigned to each Census Division based on the number of dairy cows reported by the Census of agriculture

```python
Milk_Quebec = 33478860*100 #liters

#Import animals in CDs from manure files. Dairy heifers do not produce milk
CattleManureQuebec = pd.read_csv('DataConditioning/'+'Cattle'+'_'+'Manure'+'.csv', sep=",", header=0)

DairyCowQuebec = CattleManureQuebec[['CD','DairyCow', 'DairyCow_AU']]

DairyCowTotal = DairyCowQuebec.sum(numeric_only=True)['DairyCow']
DairyCowAUTotal = DairyCowQuebec.sum(numeric_only=True)['DairyCow_AU']

#Share of cows in each CD
DairyCowQuebec['% of dairy cows (animals) in CD over total Quebec'] = DairyCowQuebec['DairyCow']/DairyCowTotal*100
DairyCowQuebec['% of dairy cows (AU) in CD over total Quebec'] = DairyCowQuebec['DairyCow_AU']/DairyCowAUTotal*100

# Assingning milk based on cows
DairyCowQuebec['Milk (liters)'] = DairyCowQuebec['% of dairy cows (animals) in CD over total Quebec']/100*Milk_Quebec

DairyCowQuebec
```

### P contained in milk

0.94 g P per L of milk (Health Canada, 2008)

https://www.canada.ca/content/dam/hc-sc/migration/hc-sc/fn-an/alt_formats/pdf/nutrition/fiche-nutri-data/nvscf-vnqau-eng.pdf

```python
P_Milk = 0.94E-3 #kg per L
```

```python
DairyCowQuebec['Milk (kg P)'] = DairyCowQuebec['Milk (liters)']*P_Milk

DairyCowQuebec.to_csv('../Results/AgriProducts/P_MilkQuebec.csv')

DairyCowQuebec
```

# Eggs


```python
Eggs = pd.read_csv('../Datasets/AgriProductsProduction/Eggs/32100376.csv', sep=",", header=0)

EggsQuebec = DGUID_Quebec.merge(Eggs, how='inner', on='DGUID')

EggsQuebec['CD'] = EggsQuebec['DGUID'].str.split('A0003').str[1]

EggsQuebec = EggsQuebec[(EggsQuebec['Unit of measure'] == 'Number of eggs (dozens)') & (EggsQuebec['Egg production'] == 'Table eggs')]

EggsQuebec = EggsQuebec.fillna(0)

EggsQuebec = EggsQuebec.rename(columns={'VALUE':'Eggs (dozens)'})

EggsQuebec['Eggs (number)'] = EggsQuebec['Eggs (dozens)']*12


EggsQuebec
```

### P contained in egg

59.5 g per egg (https://www.sciencedirect.com/science/article/pii/B9780128008799000019)

170mg P per 100 g of egg (Standardised calculation methods for animal manure and nutrients) THIS IS THE ONE USED; NOT THE ONES BELOW
    
+ 173.58 http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.1056.3102&rep=rep1&type=pdf 
    
+ 158.17 http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.1056.3102&rep=rep1&type=pdf 
    
+ 173 https://www.semanticscholar.org/paper/Effect-of-housing-systems-on-biochemical-of-chicken-Matt-Veromann/92075e20533734bde682b11d9abea77b2c679737  
    
+ 164 https://www.semanticscholar.org/paper/Effect-of-housing-systems-on-biochemical-of-chicken-Matt-Veromann/92075e20533734bde682b11d9abea77b2c679737 

+ 198 https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6470839/ 

```python
Egg_Mass = 59.5 #g/egg
PinEgg = (((170)*1E-3)/100*Egg_Mass)/1E3 #kg in egg
```

```python
EggsQuebec_summary = EggsQuebec[['DGUID','CD','Eggs (number)']]

EggsQuebec_summary['Eggs (kg P)'] = EggsQuebec_summary['Eggs (number)'] * PinEgg

EggsQuebec_summary.to_csv('../Results/AgriProducts/P_EggsQuebec.csv')


EggsQuebec_summary
```
