### Extracting Quebec subdivision levels


```python
CensusDivisions = pd.read_csv('../Datasets/GeodataFiles/CensusDivisions/georef-canada-census-division.csv', sep=";", header=0)
CensusDivisionsQuebec = CensusDivisions[CensusDivisions['Official Code Province / Territory'] == 24].reset_index(drop=True)
CensusDivisionsQuebec
```

### Importing livestock census and assign to Census divisions

### DGUID:
https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3210037001

We will consider the most recent data (2021) for the census divisions of Quebec:

Vintage: 2021

Type: A

Schema: 0003

Geographic Unique Identifier (GEOUID): Previous dataset (georef-canada-census-division.csv)

```python
DGUID_Quebec = pd.DataFrame()

DGUID_Vintage = '2021'

DGUID_Type = 'A'

DGUID_Schema =  '0003'

DGUID_Quebec['DGUID'] = DGUID_Vintage+DGUID_Type+DGUID_Schema+CensusDivisionsQuebec['Official Code Census Division'].astype(str)
DGUID_Quebec
```

```python
PoultryCensus = pd.read_csv('../Datasets/Livestock/Poultry/32100374-eng/32100374.csv', sep=",", header=0)
PoultryCensus
```

```python
PoultryCensusQuebec= DGUID_Quebec.merge(PoultryCensus, how='inner', on='DGUID')
```

Cleaning PoultryCensusQuebec Dataset

```python
# Removing 'Number of farms reporting'
PoultryCensusQuebec_Clean = PoultryCensusQuebec[PoultryCensusQuebec['Unit of measure'] != 'Number of farms reporting']

# Removing data from 2011
PoultryCensusQuebec_Clean = PoultryCensusQuebec_Clean[PoultryCensusQuebec_Clean['REF_DATE'] == 2021]

PoultryCensusQuebec_Clean['Poultry inventory'].unique()
```

```python
# Pivoting values per livestock type
PoultryCensusQuebec_Pivoted = PoultryCensusQuebec_Clean.pivot(index='DGUID', columns='Poultry inventory', values='VALUE')

# .split('/Results/Merged/')
for index, row in PoultryCensusQuebec_Pivoted.iterrows():
    PoultryCensusQuebec_Pivoted.loc[index, 'CD']= row.name.split('0003')[1]
    
PoultryCensusQuebec_Pivoted.to_csv('DataConditioning/'+'Poultry'+'_'+'CensusQuebec_Pivoted'+'.csv')
    
PoultryCensusQuebec_Pivoted
```

### Import manure composition file

```python
ManureComposition = pd.read_csv('../Datasets/ManureProperties/waste_data_input.csv', sep=",", header=0).set_index('Livestock type')

# CattleTypes = ManureComposition.index.to_list()
PoultryTypes = [s for s in ManureComposition.index.to_list() if ("Chickens" in s) or ("Turkeys" in s)]
PoultryTypes
```

### Data reconciliation between census and USDA reports

### Assumptions:

Broilers, roasters and Cornish are comprable to ChickensBroilers

Layer and broiler breeders (pullets and hens) are considered ChickensPullets

Laying hens, 19 weeks and over, that produce table eggs are considered ChickensLayers

Other poultry are comprable to ChickensBroilers

Pullets under 19 weeks, intended for laying table eggs are considered ChickensPullets

Turkeys are considered TurkeysMature

'Ducks', 'Geese', Other poultry ar enot considered

In the case of broiler, pullets and turkey the number of animals reported by the livestock census have been reduced by a factor of 0.68 and 0.80 respectively, since these animals have life cycles of 43 and 80 days respectively, maintaining the barns empty for 20 days between cycles (https://doi.org/10.4141/S06-063)

```python
PoultryCensusQuebec_PivotedClean = pd.DataFrame()
PoultryCensusQuebec_PivotedClean['CD'] = PoultryCensusQuebec_Pivoted['CD']
PoultryCensusQuebec_PivotedClean = PoultryCensusQuebec_PivotedClean.set_index('CD')

PoultryCensusQuebec_PivotedClean['ChickensBroilers'] = PoultryCensusQuebec_Pivoted.loc[:,['Broilers, roasters and Cornish', 'Other poultry']].sum(axis=1).to_numpy()*0.68

PoultryCensusQuebec_PivotedClean['ChickensPullets'] = PoultryCensusQuebec_Pivoted.loc[:,['Layer and broiler breeders (pullets and hens), total','Pullets intended for laying table eggs, under 19 weeks']].sum(axis=1).to_numpy()*0.68

PoultryCensusQuebec_PivotedClean['ChickensLayers'] = (PoultryCensusQuebec_Pivoted['Laying hens that produce table eggs, 19 weeks and over'].to_numpy())

PoultryCensusQuebec_PivotedClean['TurkeysMature'] = (PoultryCensusQuebec_Pivoted['Turkeys'].to_numpy())*0.8

PoultryCensusQuebec_PivotedClean = PoultryCensusQuebec_PivotedClean.fillna(0)

PoultryCensusQuebec_PivotedClean.to_csv('DataConditioning/'+'Poultry'+'_'+'PoultryCensusQuebec_PivotedClean'+'.csv')
```

```python
PoultryCensusQuebec_PivotedClean
```

```python
PoultryManureQuebec = PoultryCensusQuebec_PivotedClean.copy()
PoultryTypes_AU = [i + '_AU' for i in PoultryCensusQuebec_PivotedClean.columns.to_list()]
PoultryTypes_Manure = [i + '_Manure_kgperyear' for i in PoultryCensusQuebec_PivotedClean.columns.to_list()]
PoultryTypes_P = [i + '_P_kgperyear' for i in PoultryCensusQuebec_PivotedClean.columns.to_list()]
PoultryTypes_Pinorg = [i + '_Pinorg_kgperyear' for i in PoultryCensusQuebec_PivotedClean.columns.to_list()]
PoultryTypes_Porg = [i + '_Porg_kgperyear' for i in PoultryCensusQuebec_PivotedClean.columns.to_list()]


for i in PoultryTypes_AU:
    PoultryManureQuebec[i] = np.nan
PoultryManureQuebec['PoultryAnimalUnits'] = np.nan

for i in PoultryTypes_Manure:
    PoultryManureQuebec[i] = np.nan
PoultryManureQuebec['PoultryManure_kgperyear'] = np.nan

for i in PoultryTypes_P:
    PoultryManureQuebec[i] = np.nan
PoultryManureQuebec['PoultryP_kgperyear'] = np.nan

for i in PoultryTypes_Pinorg:
    PoultryManureQuebec[i] = np.nan
PoultryManureQuebec['PoultryPinorg_kgperyear'] = np.nan

for i in PoultryTypes_Porg:
    PoultryManureQuebec[i] = np.nan
PoultryManureQuebec['PoultryPorg_kgperyear'] = np.nan


for index, row in PoultryManureQuebec.iterrows():
    for i in PoultryCensusQuebec_PivotedClean.columns:
        PoultryManureQuebec.loc[index, i + '_AU'] = PoultryCensusQuebec_PivotedClean.loc[index,i]/float(ManureComposition.loc[i,'AU_equivalence'])      
    PoultryManureQuebec.loc[index,'PoultryAnimalUnits'] = sum(PoultryCensusQuebec_PivotedClean.loc[index,i]/float(ManureComposition.loc[i,'AU_equivalence']) for i in PoultryCensusQuebec_PivotedClean.columns)
    
    for i in PoultryTypes_AU:
        PoultryManureQuebec.loc[index, i.split('_AU')[0]+'_Manure_kgperyear'] = PoultryManureQuebec.loc[index, i]*float(ManureComposition.loc[i.split('_AU')[0],'Weight_kgperdayperAU'])*365
    PoultryManureQuebec.loc[index,'PoultryManure_kgperyear'] = sum(PoultryManureQuebec.loc[index, ii] for ii in PoultryTypes_Manure)
    
    for i in PoultryTypes_Manure:
        PoultryManureQuebec.loc[index, i.split('_Manure_kgperyear')[0]+'_P_kgperyear'] = PoultryManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_Manure_kgperyear')[0],'P_tot'])/100
    PoultryManureQuebec.loc[index,'PoultryP_kgperyear'] = sum(PoultryManureQuebec.loc[index, ii] for ii in PoultryTypes_P)
    
    for i in PoultryTypes_P:
        PoultryManureQuebec.loc[index, i.split('_P_kgperyear')[0]+'_Pinorg_kgperyear'] = PoultryManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_P_kgperyear')[0],'Pinorg_Ptot_ratio'])
    PoultryManureQuebec.loc[index,'PoultryPinorg_kgperyear'] = sum(PoultryManureQuebec.loc[index, ii] for ii in PoultryTypes_Pinorg)
    
    for i in PoultryTypes_P:
        PoultryManureQuebec.loc[index, i.split('_P_kgperyear')[0]+'_Porg_kgperyear'] = PoultryManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_P_kgperyear')[0],'Porg_Ptot_ratio'])
    PoultryManureQuebec.loc[index,'PoultryPorg_kgperyear'] = sum(PoultryManureQuebec.loc[index, ii] for ii in PoultryTypes_Porg)

PoultryManureQuebec
```
