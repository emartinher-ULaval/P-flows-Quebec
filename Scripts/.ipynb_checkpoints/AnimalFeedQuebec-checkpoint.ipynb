```python
PinAnimalFeed_Dataset = pd.read_csv('../Datasets/AnimalFeed/LivestockPInput.csv', sep=",", header=0).set_index('Animal')
PinAnimalFeed_Dataset
```

## Import all animal datasets


### Asumptions

Beef cows are included in the category 'Suckler, feedlot and grazing cows'

Beef calves are included in the category 'Fattening calves'

Dairy cows are included in the category 'Dairy cows'

Dairy heifers are included in the category 'Dairy young stock, 12 months and under'

Swine inmature are included in the category 'Glits and young boar'

Swine mature are included in the category 'Hogs'

Sheep are included in the category 'Sheep'

Lambs are included in the category 'Sheep'

Chicken broilers are included in the category 'Broiler'

Chicken pullets are included in the category 'Laying hen of 18 weeks and under'

Chickens layers are included in the category 'Laying hen of 18 weeks and over'

Turkeys mature are included in the category 'Turkey'

Mink are included in the category 'Minks'

Rabbits are included in the category 'Rabbits'

Goat are included in the category 'Goat'

**Horses are not accounted due to the lack of information**

**Llamas/alpacas are not accounted due to the lack of information**

**Bison are not accounted due to the lack of information**|

```python
CattleCensusQuebec = pd.read_csv('DataConditioning/'+'Cattle'+'_'+'CattleCensusQuebec_PivotedClean'+'.csv', sep=",", header=0).set_index('CD')
OtherLivestockCensusQuebec = pd.read_csv('DataConditioning/OtherLivestock_OtherLivestockCensusQuebec_PivotedClean.csv', sep=",", header=0).set_index('CD')
PoultryCensusQuebec = pd.read_csv('DataConditioning/Poultry_PoultryCensusQuebec_PivotedClean.csv', sep=",", header=0).set_index('CD')
SheepCensusQuebec = pd.read_csv('DataConditioning/Sheep_SheepCensusQuebec_PivotedClean.csv', sep=",", header=0).set_index('CD')
SwineCensusQuebec = pd.read_csv('DataConditioning/Swine_SwineCensusQuebec_PivotedClean.csv', sep=",", header=0).set_index('CD')

AnimalsCensusQuebec = pd.concat([CattleCensusQuebec, SwineCensusQuebec, SheepCensusQuebec, PoultryCensusQuebec, OtherLivestockCensusQuebec], join="inner", axis=1)

AnimalsCensusQuebec = AnimalsCensusQuebec.rename(columns=dict(zip(AnimalsCensusQuebec.columns.to_list(), [i + ' (heads)' for i in AnimalsCensusQuebec.columns.to_list()])))

AnimalsCensusQuebec
```

```python
PinAnimalFeed = pd.DataFrame()

PinAnimalFeed.loc[:,'CD'] = AnimalsCensusQuebec.index
PinAnimalFeed = PinAnimalFeed.set_index('CD')

PinAnimalFeed.loc[:,'BeefCow (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'BeefCow (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Suckler, feedlot and grazing cows','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'BeefCalf (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'BeefCalf (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Fattening calves','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'DairyCow (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'DairyCow (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Dairy cows','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'DairyHeifer (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'DairyHeifer (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Dairy young stock, 12 months and under','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'SwineInmature (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'SwineInmature (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Glits and young boar','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'SwineMature (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'SwineMature (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Hogs','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'Sheep (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'Sheep (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Sheep','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'Lamb (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'Lamb (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Sheep','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'ChickensBroilers (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'ChickensBroilers (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Broiler','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'ChickensPullets (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'ChickensPullets (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Laying hen of 18 weeks and under','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'ChickensLayers (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'ChickensLayers (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Laying hen of 18 weeks and over','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'TurkeysMature (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'TurkeysMature (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Turkey','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'Mink (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'Mink (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Minks','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'Rabbits (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'Rabbits (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Rabbits','Uptake P (kg/animal/year)'])

PinAnimalFeed.loc[:,'Goat (kg P feed/year)'] = (AnimalsCensusQuebec.loc[:,'Goat (heads)']*
                                                  PinAnimalFeed_Dataset.loc['Goat','Uptake P (kg/animal/year)'])

PinAnimalFeed.to_csv('../Results/AnimalFeed/PinAnimalFeedCleanQuebec.csv')


PinAnimalFeed
```

```python
PinAnimalFeed_Summary = pd.DataFrame()

PinAnimalFeed_Summary.loc[:,'CD'] = PinAnimalFeed.index
PinAnimalFeed_Summary = PinAnimalFeed_Summary.set_index('CD')

PinAnimalFeed_Summary['2021 kg P feed/year'] = PinAnimalFeed.sum(axis=1)

PinAnimalFeed_Summary.to_csv('../Results/AnimalFeed/PinAnimalFeedQuebec_Summary.csv')

PinAnimalFeed_Summary
```
