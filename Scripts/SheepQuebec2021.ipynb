```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import glob
# import seaborn as sns
# %matplotlib notebook
%matplotlib inline

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

### Extracting Quebec subdivision levels


```python
CensusDivisions = pd.read_csv('../Datasets/GeodataFiles/CensusDivisions/georef-canada-census-division.csv', sep=";", header=0)
CensusDivisionsQuebec = CensusDivisions[CensusDivisions['Official Code Province / Territory'] == 24].reset_index(drop=True)
CensusDivisionsQuebec
```

### Importing livestock census and assign to Census divisions

### DGUID:
https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3210037001

We will consider the most recent data (2021) for the census divisions of Quebec:

Vintage: 2021

Type: A

Schema: 0003

Geographic Unique Identifier (GEOUID): Previous dataset (georef-canada-census-division.csv)

```python
DGUID_Quebec = pd.DataFrame()

DGUID_Vintage = '2021'

DGUID_Type = 'A'

DGUID_Schema =  '0003'

DGUID_Quebec['DGUID'] = DGUID_Vintage+DGUID_Type+DGUID_Schema+CensusDivisionsQuebec['Official Code Census Division'].astype(str)
DGUID_Quebec
```

```python
SheepCensus = pd.read_csv('../Datasets/Livestock/Sheep/32100371-eng/32100371.csv', sep=",", header=0)
SheepCensus
```

```python
SheepCensusQuebec = DGUID_Quebec.merge(SheepCensus, how='inner', on='DGUID')
```

Cleaning SheepCensusQuebec Dataset

```python
# Removing 'Number of farms reporting'
SheepCensusQuebec_Clean = SheepCensusQuebec[SheepCensusQuebec['Unit of measure'] != 'Number of farms reporting']

# Removing data from 2011
SheepCensusQuebec_Clean = SheepCensusQuebec_Clean[SheepCensusQuebec_Clean['REF_DATE'] == 2021]

SheepCensusQuebec_Clean['Sheep'].unique()
```

```python
# Pivoting values per livestock type
SheepCensusQuebec_Pivoted = SheepCensusQuebec_Clean.pivot(index='DGUID', columns='Sheep', values='VALUE')

# .split('/Results/Merged/')
for index, row in SheepCensusQuebec_Pivoted.iterrows():
    SheepCensusQuebec_Pivoted.loc[index, 'CD']= row.name.split('0003')[1]
    
SheepCensusQuebec_Pivoted.to_csv('DataConditioning/'+'Sheep'+'_'+'CensusQuebec_Pivoted'+'.csv')
    
SheepCensusQuebec_Pivoted
```

### Import manure composition file

```python
ManureComposition = pd.read_csv('../Datasets/ManureProperties/waste_data_input.csv', sep=",", header=0).set_index('Livestock type')

# CattleTypes = ManureComposition.index.to_list()
SheepTypes = [s for s in ManureComposition.index.to_list() if ("Sheep" in s) or ("Lamb" in s)]
SheepTypes
```

### Data reconciliation between census and USDA reports

### Assumptions:

Ewes and rams are comparable to sheeps

Lambs are lambs

```python
SheepCensusQuebec_PivotedClean = pd.DataFrame()
SheepCensusQuebec_PivotedClean['CD'] = SheepCensusQuebec_Pivoted['CD']
SheepCensusQuebec_PivotedClean = SheepCensusQuebec_PivotedClean.set_index('CD')

SheepCensusQuebec_PivotedClean['Sheep'] = SheepCensusQuebec_Pivoted.loc[:,['Ewes','Rams']].sum(axis=1).to_numpy()

SheepCensusQuebec_PivotedClean['Lamb'] = (SheepCensusQuebec_Pivoted['Lambs, total'].to_numpy())

SheepCensusQuebec_PivotedClean = SheepCensusQuebec_PivotedClean.fillna(0)

SheepCensusQuebec_PivotedClean.to_csv('DataConditioning/'+'Sheep'+'_'+'SheepCensusQuebec_PivotedClean'+'.csv')
```

```python
SheepManureQuebec = SheepCensusQuebec_PivotedClean.copy()
SheepTypes_AU = [i + '_AU' for i in SheepCensusQuebec_PivotedClean.columns.to_list()]
SheepTypes_Manure = [i + '_Manure_kgperyear' for i in SheepCensusQuebec_PivotedClean.columns.to_list()]
SheepTypes_P = [i + '_P_kgperyear' for i in SheepCensusQuebec_PivotedClean.columns.to_list()]
SheepTypes_Pinorg = [i + '_Pinorg_kgperyear' for i in SheepCensusQuebec_PivotedClean.columns.to_list()]
SheepTypes_Porg = [i + '_Porg_kgperyear' for i in SheepCensusQuebec_PivotedClean.columns.to_list()]


for i in SheepTypes_AU:
    SheepManureQuebec[i] = np.nan
SheepManureQuebec['SheepAnimalUnits'] = np.nan

for i in SheepTypes_Manure:
    SheepManureQuebec[i] = np.nan
SheepManureQuebec['SheepManure_kgperyear'] = np.nan

for i in SheepTypes_P:
    SheepManureQuebec[i] = np.nan
SheepManureQuebec['SheepP_kgperyear'] = np.nan

for i in SheepTypes_Pinorg:
    SheepManureQuebec[i] = np.nan
SheepManureQuebec['SheepPinorg_kgperyear'] = np.nan

for i in SheepTypes_Porg:
    SheepManureQuebec[i] = np.nan
SheepManureQuebec['SheepPorg_kgperyear'] = np.nan


for index, row in SheepManureQuebec.iterrows():
    for i in SheepCensusQuebec_PivotedClean.columns:
        SheepManureQuebec.loc[index, i + '_AU'] = SheepCensusQuebec_PivotedClean.loc[index,i]/float(ManureComposition.loc[i,'AU_equivalence'])
    SheepManureQuebec.loc[index,'SheepAnimalUnits'] = sum(SheepCensusQuebec_PivotedClean.loc[index,i]/float(ManureComposition.loc[i,'AU_equivalence']) for i in SheepCensusQuebec_PivotedClean.columns)
    
    for i in SheepTypes_AU:
        SheepManureQuebec.loc[index, i.split('_AU')[0]+'_Manure_kgperyear'] = SheepManureQuebec.loc[index, i]*float(ManureComposition.loc[i.split('_AU')[0],'Weight_kgperdayperAU'])*365
    SheepManureQuebec.loc[index,'SheepManure_kgperyear'] = sum(SheepManureQuebec.loc[index, ii] for ii in SheepTypes_Manure)
    
    for i in SheepTypes_Manure:
        SheepManureQuebec.loc[index, i.split('_Manure_kgperyear')[0]+'_P_kgperyear'] = SheepManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_Manure_kgperyear')[0],'P_tot'])/100
    SheepManureQuebec.loc[index,'SheepP_kgperyear'] = sum(SheepManureQuebec.loc[index, ii] for ii in SheepTypes_P)
    
    for i in SheepTypes_P:
        SheepManureQuebec.loc[index, i.split('_P_kgperyear')[0]+'_Pinorg_kgperyear'] = SheepManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_P_kgperyear')[0],'Pinorg_Ptot_ratio'])
    SheepManureQuebec.loc[index,'SheepPinorg_kgperyear'] = sum(SheepManureQuebec.loc[index, ii] for ii in SheepTypes_Pinorg)
    
    for i in SheepTypes_P:
        SheepManureQuebec.loc[index, i.split('_P_kgperyear')[0]+'_Porg_kgperyear'] = SheepManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_P_kgperyear')[0],'Porg_Ptot_ratio'])
    SheepManureQuebec.loc[index,'SheepPorg_kgperyear'] = sum(SheepManureQuebec.loc[index, ii] for ii in SheepTypes_Porg)

SheepManureQuebec
```

```python
SheepManureQuebec.to_csv('DataConditioning/'+'Sheep'+'_'+'ManureQuebec'+'.csv')
```
