### Extracting Quebec subdivision levels


```python
CensusDivisions = pd.read_csv('../Datasets/GeodataFiles/CensusDivisions/georef-canada-census-division.csv', sep=";", header=0)
CensusDivisionsQuebec = CensusDivisions[CensusDivisions['Official Code Province / Territory'] == 24].reset_index(drop=True)
CensusDivisionsQuebec
```

### Importing livestock census and assign to Census divisions

### DGUID:
https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3210037001

We will consider the most recent data (2021) for the census divisions of Quebec:

Vintage: 2021

Type: A

Schema: 0003

Geographic Unique Identifier (GEOUID): Previous dataset (georef-canada-census-division.csv)

```python
DGUID_Quebec = pd.DataFrame()

DGUID_Vintage = '2021'

DGUID_Type = 'A'

DGUID_Schema =  '0003'

DGUID_Quebec['DGUID'] = DGUID_Vintage+DGUID_Type+DGUID_Schema+CensusDivisionsQuebec['Official Code Census Division'].astype(str)
DGUID_Quebec
```

```python
SwineCensus = pd.read_csv('../Datasets/Livestock/Pig/32100372-eng/32100372.csv', sep=",", header=0)
SwineCensus
```

```python
SwineCensusQuebec = DGUID_Quebec.merge(SwineCensus, how='inner', on='DGUID')
SwineCensusQuebec['Pigs'].unique()
# Check if the data is well extracted
# CattleCensusQuebec[(CattleCensusQuebec['Cattle and calves']=='Total cattle and calves') & (CattleCensusQuebec['Unit of measure']=='Number of animals') & (CattleCensusQuebec['REF_DATE']==2016)].count()
```

Cleaning SwineCensus Dataset

```python
# Removing 'Number of farms reporting'
SwineCensusQuebec_Clean = SwineCensusQuebec[SwineCensusQuebec['Unit of measure'] != 'Number of farms reporting']

# Removing data from 2011
SwineCensusQuebec_Clean = SwineCensusQuebec_Clean[SwineCensusQuebec_Clean['REF_DATE'] == 2021]

SwineCensusQuebec_Clean['Pigs'].unique()
```

```python
# Pivoting values per livestock type
SwineCensusQuebec_Pivoted = SwineCensusQuebec_Clean.pivot(index='DGUID', columns='Pigs', values='VALUE')

# .split('/Results/Merged/')
for index, row in SwineCensusQuebec_Pivoted.iterrows():
    SwineCensusQuebec_Pivoted.loc[index, 'CD']= row.name.split('0003')[1]
    
SwineCensusQuebec_Pivoted.to_csv('DataConditioning/'+'Swine'+'_'+'CensusQuebec_Pivoted'+'.csv')
    
SwineCensusQuebec_Pivoted
```

### Data reconciliation between census and USDA reports

### Assumptions:

Grower and finishing pigs: Inmature swine

Nursing pigs: Inmature swine

Sows and gilts for breeding: Mature swine

Boars: Mature swine

Weaner pigs: Immature swine

```python
SwineCensusQuebec_Pivoted.columns
```

```python
SwineCensusQuebec_PivotedClean = pd.DataFrame()
SwineCensusQuebec_PivotedClean['CD'] = SwineCensusQuebec_Pivoted['CD']
SwineCensusQuebec_PivotedClean = SwineCensusQuebec_PivotedClean.set_index('CD')

if 'Nursing and weaner pigs' in SwineCensusQuebec_Pivoted.columns: #'Nursing and weaner pigs' is not reported for 2016
    SwineCensusQuebec_PivotedClean['SwineInmature'] = SwineCensusQuebec_Pivoted.loc[:,['Market pigs, 50 lb (23 kg) and over',
                                                                                      'Suckling pigs, less than 15 lb (7 kg)', 'Weanling, nursery or starter pigs, 15 to 49 lb (7 to 22 kg)']].sum(axis=1).to_numpy()
else:
    SwineCensusQuebec_PivotedClean['SwineInmature'] = SwineCensusQuebec_Pivoted.loc[:,['Market pigs, 50 lb (23 kg) and over',
                                                                                      'Suckling pigs, less than 15 lb (7 kg)', 'Weanling, nursery or starter pigs, 15 to 49 lb (7 to 22 kg)',]].sum(axis=1).to_numpy()
    
SwineCensusQuebec_PivotedClean['SwineMature'] = SwineCensusQuebec_Pivoted.loc[:,['Sows and gilts for breeding','Boars for breeding']].sum(axis=1).to_numpy()


SwineCensusQuebec_PivotedClean = SwineCensusQuebec_PivotedClean.fillna(0)

SwineCensusQuebec_PivotedClean.to_csv('DataConditioning/'+'Swine'+'_'+'SwineCensusQuebec_PivotedClean'+'.csv')
```

```python
SwineCensusQuebec_PivotedClean
```

```python
SwineManureQuebec = SwineCensusQuebec_PivotedClean.copy()
SwineTypes_AU = [i + '_AU' for i in SwineCensusQuebec_PivotedClean.columns.to_list()]
SwineTypes_Manure = [i + '_Manure_kgperyear' for i in SwineCensusQuebec_PivotedClean.columns.to_list()]
SwineTypes_P = [i + '_P_kgperyear' for i in SwineCensusQuebec_PivotedClean.columns.to_list()]
SwineTypes_Pinorg = [i + '_Pinorg_kgperyear' for i in SwineCensusQuebec_PivotedClean.columns.to_list()]
SwineTypes_Porg = [i + '_Porg_kgperyear' for i in SwineCensusQuebec_PivotedClean.columns.to_list()]


for i in SwineTypes_AU:
    SwineManureQuebec[i] = np.nan
SwineManureQuebec['SwineAnimalUnits'] = np.nan

for i in SwineTypes_Manure:
    SwineManureQuebec[i] = np.nan
SwineManureQuebec['SwineManure_kgperyear'] = np.nan

for i in SwineTypes_P:
    SwineManureQuebec[i] = np.nan
SwineManureQuebec['SwineP_kgperyear'] = np.nan

for i in SwineTypes_Pinorg:
    SwineManureQuebec[i] = np.nan
SwineManureQuebec['SwinePinorg_kgperyear'] = np.nan

for i in SwineTypes_Porg:
    SwineManureQuebec[i] = np.nan
SwineManureQuebec['SwinePorg_kgperyear'] = np.nan


for index, row in SwineManureQuebec.iterrows():
    for i in SwineCensusQuebec_PivotedClean.columns:
        SwineManureQuebec.loc[index, i + '_AU'] = SwineCensusQuebec_PivotedClean.loc[index,i]/float(ManureComposition.loc[i,'AU_equivalence'])
    SwineManureQuebec.loc[index,'SwineAnimalUnits'] = sum(SwineCensusQuebec_PivotedClean.loc[index,i]/float(ManureComposition.loc[i,'AU_equivalence']) for i in SwineCensusQuebec_PivotedClean.columns)
    
    for i in SwineTypes_AU:
        SwineManureQuebec.loc[index, i.split('_AU')[0]+'_Manure_kgperyear'] = SwineManureQuebec.loc[index, i]*float(ManureComposition.loc[i.split('_AU')[0],'Weight_kgperdayperAU'])*365
    SwineManureQuebec.loc[index,'SwineManure_kgperyear'] = sum(SwineManureQuebec.loc[index, ii] for ii in SwineTypes_Manure)
    
    for i in SwineTypes_Manure:
        SwineManureQuebec.loc[index, i.split('_Manure_kgperyear')[0]+'_P_kgperyear'] = SwineManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_Manure_kgperyear')[0],'P_tot'])/100
    SwineManureQuebec.loc[index,'SwineP_kgperyear'] = sum(SwineManureQuebec.loc[index, ii] for ii in SwineTypes_P)
    
    for i in SwineTypes_P:
        SwineManureQuebec.loc[index, i.split('_P_kgperyear')[0]+'_Pinorg_kgperyear'] = SwineManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_P_kgperyear')[0],'Pinorg_Ptot_ratio'])
    SwineManureQuebec.loc[index,'SwinePinorg_kgperyear'] = sum(SwineManureQuebec.loc[index, ii] for ii in SwineTypes_Pinorg)
    
    for i in SwineTypes_P:
        SwineManureQuebec.loc[index, i.split('_P_kgperyear')[0]+'_Porg_kgperyear'] = SwineManureQuebec.loc[index, i]*(ManureComposition.loc[i.split('_P_kgperyear')[0],'Porg_Ptot_ratio'])
    SwineManureQuebec.loc[index,'SwinePorg_kgperyear'] = sum(SwineManureQuebec.loc[index, ii] for ii in SwineTypes_Porg)

SwineManureQuebec
```
